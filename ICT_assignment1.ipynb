Question 1 (Reads iris dataset to data frame 'data_iris',
then prints the data frame with its summary, first and last 5 rows, etc)

```python
import pandas as pd
data_iris=pd.read_excel('/content/iris.xls')
#\033[1;4m for switching the text to bold and underline mode
#\033[0m for switching back to normal mode
print("\033[1;4mEntries in Iris data set:\033[0m")
print(data_iris,'\n')
print("\033[1;4mA brief summary of Iris data set:\033[0m")
print(data_iris.info(),'\n')
print("\033[1;4mStatistical summary of Iris data set:\033[0m")
print(data_iris.describe(),'\n')
print("\033[1;4mFirst five rows in Iris data set:\033[0m")
print(data_iris.head(),'\n')
print("\033[1;4mLast five rows in Iris data set:\033[0m")
print(data_iris.tail(),'\n')
```

Question 2 (displays the names of all columns and then displays values in each columns as a series)

```python
import pandas as pd
data_iris=pd.read_excel('/content/iris.xls')
col_names=data_iris.keys()  #to fetch the column names from DataFrame and store it in col_names
print("Name of columns in the dataset are:")
for  i in col_names: #this for loop displays names of all columns
  print(i)
print('\n')

for i in col_names: #this for loop displays values in each columns
  print('Values in column', i, 'are:\n')
  print(data_iris[i])
  print('\n')
```

Question 3 (columns with numerical data types are identified first and then mean of only those colums are calculated)

```python
import pandas as pd
data_iris=pd.read_excel('/content/iris.xls')
col_names=data_iris.keys()  #to fetch the column names from DataFrame and store it in col_names
#this for loop iterates through all columns and checks if the data type of that column is numeric(int, float, etc)
#if yes, then mean of that column is calculated
for i in col_names:
  if(data_iris[i].dtype!='O'): # this if condition checks whether the entries in the prticular column is of numerical data type
    print('Mean of column', i, 'is: ',data_iris[i].mean())
```

Question 5 (visualizations are made using three kind of plots: scatter plot,
box plot and count plot. Subplots are used in each of above plots)

```python
import pandas as pd
data_iris=pd.read_excel('/content/iris.xls')
import matplotlib.pyplot as plt
import seaborn as sns

fig=plt.figure(figsize=(15,7)) # to set figure size

plt.suptitle('Scatter plot of SL vs SW and PL vs PW for each species');
plt.subplot(1,2,1) #first scatter subplot, viz, SL vs SW
sns.scatterplot(x='SL',y='SW',data=data_iris,hue='Classification');

plt.subplot(1,2,2) #second scatter subplot, viz, PL vs PW
sns.scatterplot(x='PL',y='PW',data=data_iris,hue='Classification');
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_iris=pd.read_excel('/content/iris.xls')

fig=plt.figure(figsize=(20,10))# to set figure size

plt.suptitle('Box plot of SL,SW,PL and PW for each species');
plt.subplot(2,2,1) #first box subplot for SL
sns.boxplot(y='SL',data=data_iris,hue='Classification',showmeans=True);

plt.subplot(2,2,2) #second box subplot for SW
sns.boxplot(y='SW',data=data_iris,hue='Classification',showmeans=True);

plt.subplot(2,2,3) #third box subplot for PL
sns.boxplot(y='PL',data=data_iris,hue='Classification',showmeans=True);

plt.subplot(2,2,4) #fourth box subplot for PW
sns.boxplot(y='PW',data=data_iris,hue='Classification',showmeans=True);

plt.subplots_adjust(top=0.95) #this will adjust the spacing between suptitle and subplots
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data_iris=pd.read_excel('/content/iris.xls')

fig=plt.figure(figsize=(20,10)) # to set figure size

plt.suptitle('Count plot of SL,SW,PL and PW for each species');
plt.subplot(2,2,1) #first count subplot for SL
sns.countplot(x='SL',data=data_iris,hue='Classification');
plt.xticks(rotation=90);

plt.subplot(2,2,2) #second count subplot for SW
sns.countplot(x='SW',data=data_iris,hue='Classification');
plt.xticks(rotation=90);

plt.subplot(2,2,3) #third count subplot for PL
sns.countplot(x='PL',data=data_iris,hue='Classification');
plt.xticks(rotation=90);

plt.subplot(2,2,4) #fourth count subplot for PW
sns.countplot(x='PW',data=data_iris,hue='Classification');
plt.xticks(rotation=90);
```
